```python
import numpy as np
import torch
from torch import nn
from torch.utils.data import DataLoader, TensorDataset
from scipy.special import logit, expit

import matplotlib.pyplot as plt
import corner

# pip install nflows
from nflows.distributions import StandardNormal
from nflows.flows import Flow
from nflows.transforms import CompositeTransform
from nflows.transforms.autoregressive import MaskedAffineAutoregressiveTransform
from nflows.transforms.permutations import ReversePermutation

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ============================================================
# 1. GW150914-LIKE POSTERIOR SAMPLES
# ============================================================

import requests
import h5py
import io
import numpy as np
import os

# ------------------------------------------------------------
# 1. Download GW150914 PE samples (GWTC-2.1)
# ------------------------------------------------------------
filename = "IGWN-GWTC2p1-v2-GW150914_095045_PEDataRelease_mixed_nocosmo.h5"
url = "https://zenodo.org/records/6513631/files/IGWN-GWTC2p1-v2-GW150914_095045_PEDataRelease_mixed_nocosmo.h5?download=1"

if not os.path.exists(filename):
    print("Downloading PE file with wget...")
    !wget -O {filename} "{url}"
else:
    print("File already exists.")

print("Loaded, parsing HDF5...")
f = h5py.File(filename, "r")

# Inspect available datasets
print("Datasets:", list(f.keys()))

# ------------------------------------------------------------
# 2. Extract relevant parameters
# ------------------------------------------------------------
# Access parameters from the 'C01:Mixed/posterior_samples' group
data_group = f['C01:Mixed']['posterior_samples']
m1_src = data_group["mass_1_source"][:]          # source-frame masses
m2_src = data_group["mass_2_source"][:]
z      = data_group["redshift"][:]               # redshift
DL     = data_group["luminosity_distance"][:]    # Mpc
ra     = data_group["ra"][:]        # right ascension
dec    = data_group["dec"][:]            # declination

# ------------------------------------------------------------
# 3. Convert to detector-frame masses
# ------------------------------------------------------------
m1_det = m1_src * (1 + z)
m2_det = m2_src * (1 + z)

# ------------------------------------------------------------
# 4. Enforce mass ordering convention m1 >= m2
# ------------------------------------------------------------
m1 = np.maximum(m1_det, m2_det)
m2 = np.minimum(m1_det, m2_det)

samples = np.column_stack([m1, m2, DL, ra, dec])

print("Final sample shape:", samples.shape)
print("Example rows:\n", samples[:5])


# ============================================================
# 2. PREPROCESSING: Transforms to Unbounded Space + Whitening
# ============================================================

# Function to transform physical params to unbounded space
def forward_transform(samples):
    x = samples.copy()
    # 1. Masses & Distance: Log transform (since > 0)
    x[:, :3] = np.log(x[:, :3])

    # 2. RA: Periodic [0, 2pi]. Map to Real using Logit on scaled variable.
    # We scale alpha to (0, 1) first.
    # Note: If samples touch boundaries exactly, apply clipping or assume they don't.
    x[:, 3] = logit(x[:, 3] / (2 * np.pi))

    # 3. Dec: [-pi/2, pi/2]. Prior is uniform in sin(dec).
    # Map delta -> sin(delta) in [-1, 1] -> scale to (0, 1) -> Logit
    s = np.sin(x[:, 4])
    s_scaled = (s + 1) / 2
    x[:, 4] = logit(s_scaled)

    return x

x_transform = forward_transform(samples)

# Whitening
mu = x_transform.mean(axis=0)
cov = np.cov(x_transform.T)
L = np.linalg.cholesky(cov)

x_white = np.linalg.solve(L, (x_transform - mu).T).T
x_white = x_white.astype(np.float32)

dataset = TensorDataset(torch.from_numpy(x_white))
loader = DataLoader(dataset, batch_size=2048, shuffle=True, drop_last=True)

dim = 5

# ============================================================
# 3. DEFINE NORMALIZING FLOW (MAF)
# ============================================================

def build_maf_flow(dim, n_layers=6, hidden_features=128):
    transforms = []
    for _ in range(n_layers):
        transforms.append(
            MaskedAffineAutoregressiveTransform(
                features=dim,
                hidden_features=hidden_features,
                num_blocks=2,
                use_residual_blocks=True,
                activation=nn.ReLU(),
            )
        )
        transforms.append(ReversePermutation(features=dim))

    transform = CompositeTransform(transforms)
    base_dist = StandardNormal([dim])
    return Flow(transform, base_dist)

flow = build_maf_flow(dim).to(device)
optimizer = torch.optim.Adam(flow.parameters(), lr=1e-3)

# ============================================================
# 4. TRAINING LOOP
# ============================================================

n_epochs = 150
flow.train()

for epoch in range(1, n_epochs + 1):
    total_loss = 0.0
    for (batch,) in loader:
        batch = batch.to(device)
        loss = -flow.log_prob(batch).mean()

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item() * batch.size(0)

    total_loss /= len(dataset)
    if epoch % 10 == 0:
        print(f"Epoch {epoch:4d} | NLL = {total_loss:.4f}")

# ============================================================
# 5. SAMPLE FROM FLOW AND INVERSE TRANSFORM
# ============================================================

def inverse_transform(x_white, mu, L):
    # 1. Unwhiten
    x_transform = (L @ x_white.T).T + mu
    x_rec = np.empty_like(x_transform)

    # 2. Inverse Log (Masses, Dist)
    x_rec[:, :3] = np.exp(x_transform[:, :3])

    # 3. Inverse RA: Sigmoid -> Scale to [0, 2pi]
    x_rec[:, 3] = expit(x_transform[:, 3]) * 2 * np.pi

    # 4. Inverse Dec: Sigmoid -> Scale to [-1, 1] (sin dec) -> arcsin
    # Clip to avoid numerical errors slightly outside [-1, 1]
    s = expit(x_transform[:, 4]) * 2 - 1
    s = np.clip(s, -1.0 + 1e-7, 1.0 - 1e-7)
    x_rec[:, 4] = np.arcsin(s)

    return x_rec

flow.eval()
with torch.no_grad():
    # Sample the same number of points as the original dataset
    z = flow.sample(len(samples)).cpu().numpy()

x_rec = inverse_transform(z, mu, L)

m1_rec = x_rec[:, 0]
m2_rec = x_rec[:, 1]
DL_rec = x_rec[:, 2]
ra_rec = x_rec[:, 3]
dec_rec = x_rec[:, 4]

# ============================================================
# 6. DIAGNOSTIC CORNER PLOTS
# ============================================================

labels = [r"$m_{1,\rm det}$", r"$m_{2,\rm det}$", r"$D_L$", r"$\alpha$", r"$\delta$"]

fig = corner.corner(
    samples,
    labels=labels,
    color="C0",
    bins=40,
    plot_datapoints=False,
    smooth=1.0,
    label_kwargs={"fontsize": 14},
)

corner.corner(
    x_rec,
    fig=fig,
    color="C3",
    bins=40,
    plot_datapoints=False,
    smooth=1.0,
)

plt.suptitle("GW150914 Mock Posterior vs Flow Samples", fontsize=16)
plt.show()
```

```python
def evaluate_logpdfs(physical_samples_input):
    # Ensure input is 2D (N_samples, N_dims)
    if physical_samples_input.ndim == 1:
        physical_samples_input = physical_samples_input.reshape(1, -1)

    # 1. Apply forward_transform
    transformed_samples = forward_transform(physical_samples_input)

    # 2. Whiten the transformed samples
    # 'L' and 'mu' are globally available from the preprocessing step
    whitened_samples = np.linalg.solve(L, (transformed_samples - mu).T).T

    # 3. Convert to PyTorch tensor and move to device
    whitened_samples_tensor = torch.from_numpy(whitened_samples.astype(np.float32)).to(device)

    # 4. Evaluate log_prob
    flow.eval() # Ensure flow is in evaluation mode
    with torch.no_grad():
        log_probs = flow.log_prob(whitened_samples_tensor).cpu().numpy()

    return log_probs

# --- Example Usage ---

# Evaluate logpdfs for a few of the original samples
example_physical_samples = samples[:5] # Using the first 5 original samples
log_likelihoods = evaluate_logpdfs(example_physical_samples)

print("Log-probabilities for example physical samples:")
for i, log_prob in enumerate(log_likelihoods):
    # Note: These are log-probabilities, not log-likelihoods in a statistical sense
    # unless the flow represents the true likelihood function. For a posterior, it's log-posterior density.
    print(f"Sample {i+1}: {log_prob:.4f}")

# Evaluate for a single hypothetical point
# Example point (m1, m2, DL, ra, dec) - adjust as needed
single_point = np.array([40.0, 30.0, 400.0, 1.5, -1.2])
single_point_logprob = evaluate_logpdfs(single_point)
print(f"\nLog-probability for a single hypothetical point {single_point}: {single_point_logprob[0]:.4f}")
```

### Quantitative Evaluation: Kolmogorov-Smirnov (KS) Test

The Kolmogorov-Smirnov test is a non-parametric test that compares the empirical cumulative distribution functions (ECDFs) of two samples. We'll perform it for each 1D marginal distribution to check if the original and flow-generated samples come from the same underlying distribution. A small p-value (e.g., < 0.05) would suggest that the distributions are statistically significantly different.

```python
from scipy import stats

print("--- Kolmogorov-Smirnov Test for 1D Marginals ---")
print("H0: The two samples are drawn from the same continuous distribution.")
print("If p-value is small (e.g., < 0.05), we reject H0, meaning distributions are likely different.")

labels = [r"$m_{1,\rm det}$", r"$m_{2,\rm det}$", r"$D_L$", r"$\alpha$", r"$\delta$"]

for i, label in enumerate(labels):
    stat, p_value = stats.kstest(samples[:, i], x_rec[:, i])
    print(f"\nParameter: {label}")
    print(f"  KS Statistic: {stat:.4f}")
    print(f"  P-value: {p_value:.4f}")
    if p_value < 0.05:
        print("  -> Distributions are statistically significantly different (reject H0)")
    else:
        print("  -> Distributions are not statistically significantly different (fail to reject H0)")
```

```python
# ============================================================
# 0. INSTALLS (if needed)
# ============================================================
# pip install flowjax optax equinox jax jaxlib corner h5py

import os
import h5py
import numpy as np
from scipy.special import logit, expit

import matplotlib.pyplot as plt
import corner

import jax
import jax.numpy as jnp
import optax
import equinox as eqx

# FlowJAX (new API)
from flowjax.flows import masked_autoregressive_flow
from flowjax.distributions import Normal

# ============================================================
# 1. LOAD GW150914 PE SAMPLES
# ============================================================

filename = "IGWN-GWTC2p1-v2-GW150914_095045_PEDataRelease_mixed_nocosmo.h5"
url = "https://zenodo.org/records/6513631/files/IGWN-GWTC2p1-v2-GW150914_095045_PEDataRelease_mixed_nocosmo.h5?download=1"

if not os.path.exists(filename):
    print("Downloading PE file...")
    os.system(f"wget -O {filename} '{url}'")
else:
    print("File already exists.")

print("Parsing HDF5...")
f = h5py.File(filename, "r")
data_group = f["C01:Mixed"]["posterior_samples"]

m1_src = data_group["mass_1_source"][:]
m2_src = data_group["mass_2_source"][:]
z      = data_group["redshift"][:]
DL     = data_group["luminosity_distance"][:]
ra     = data_group["ra"][:]
dec    = data_group["dec"][:]

m1_det = m1_src * (1 + z)
m2_det = m2_src * (1 + z)

m1 = np.maximum(m1_det, m2_det)
m2 = np.minimum(m1_det, m2_det)

samples = np.column_stack([m1, m2, DL, ra, dec])
print("Samples shape:", samples.shape)

# ============================================================
# 2. FORWARD / INVERSE TRANSFORMS + WHITENING
# ============================================================

def forward_transform(samples_np: np.ndarray) -> np.ndarray:
    x = samples_np.copy()
    x[:, :3] = np.log(x[:, :3])
    x[:, 3] = logit(x[:, 3] / (2 * np.pi))
    s = np.sin(x[:, 4])
    x[:, 4] = logit((s + 1) / 2)
    return x

def inverse_transform(x_white_np: np.ndarray, mu_np: np.ndarray, L_np: np.ndarray) -> np.ndarray:
    x_transform = (L_np @ x_white_np.T).T + mu_np
    x_rec = np.empty_like(x_transform)
    x_rec[:, :3] = np.exp(x_transform[:, :3])
    x_rec[:, 3] = expit(x_transform[:, 3]) * 2 * np.pi
    s = expit(x_transform[:, 4]) * 2 - 1
    s = np.clip(s, -1 + 1e-7, 1 - 1e-7)
    x_rec[:, 4] = np.arcsin(s)
    return x_rec

x_transform = forward_transform(samples)
mu = x_transform.mean(axis=0)
cov = np.cov(x_transform.T)
L = np.linalg.cholesky(cov)

x_white = np.linalg.solve(L, (x_transform - mu).T).T.astype(np.float32)
N, dim = x_white.shape
print("Whitened shape:", x_white.shape)

x_white_jax = jnp.asarray(x_white)

# ============================================================
# 3. BUILD FLOWJAX MAF (new API)
# ============================================================

key = jax.random.key(0)

base_dist = Normal(jnp.zeros(dim), jnp.ones(dim))

flow = masked_autoregressive_flow(
    key=key,
    base_dist=base_dist,
    flow_layers=6,
    nn_width=128,
    nn_depth=2,
)

# ============================================================
# 4. TRAINING LOOP (Equinox + Optax)
# ============================================================

batch_size = 2048
n_epochs = 500
optimizer = optax.adam(1e-3)

# Only trainable leaves (arrays)
opt_state = optimizer.init(eqx.filter(flow, eqx.is_array))

@eqx.filter_value_and_grad
def loss_fn(flow, x):
    return -jnp.mean(flow.log_prob(x))

@eqx.filter_jit
def train_step(flow, opt_state, x):
    loss, grads = loss_fn(flow, x)
    updates, opt_state = optimizer.update(grads, opt_state)
    flow = eqx.apply_updates(flow, updates)
    return flow, opt_state, loss

for epoch in range(1, n_epochs + 1):
    perm = np.random.permutation(N)
    x_perm = x_white_jax[perm]

    losses = []
    for i in range(0, N, batch_size):
        batch = x_perm[i:i+batch_size]
        if batch.shape[0] < batch_size:
            continue
        flow, opt_state, loss = train_step(flow, opt_state, batch)
        losses.append(float(loss))

    if epoch % 10 == 0:
        print(f"Epoch {epoch:4d} | NLL = {np.mean(losses):.4f}")

# ============================================================
# 5. SAMPLING FROM THE FLOW
# ============================================================

key, subkey = jax.random.split(key)
z_samples = np.array(flow.sample(subkey, (N,)))

# ============================================================
# 6. INVERSE TRANSFORM BACK TO PHYSICAL SPACE
# ============================================================

x_rec = inverse_transform(z_samples, mu, L)

# ============================================================
# 7. DIAGNOSTIC CORNER PLOT
# ============================================================

labels = [r"$m_{1,\rm det}$", r"$m_{2,\rm det}$", r"$D_L$", r"$\alpha$", r"$\delta$"]

fig = corner.corner(
    samples,
    labels=labels,
    color="C0",
    bins=40,
    plot_datapoints=False,
    smooth=1.0,
    label_kwargs={"fontsize": 14},
)

corner.corner(
    x_rec,
    fig=fig,
    color="C3",
    bins=40,
    plot_datapoints=False,
    smooth=1.0,
)

plt.suptitle("GW150914: Real Posterior vs FlowJAX MAF (new API)", fontsize=16)
plt.show()
```

```python
import numpy as np
import jax
import jax.numpy as jnp
import jax.scipy as jsp
import equinox as eqx
import optax
import dynesty
from dynesty import NestedSampler
import matplotlib.pyplot as plt
import corner

from flowjax.flows import masked_autoregressive_flow
from flowjax.distributions import Normal

# =========================================================
# Simple cosmology (toy)
# =========================================================

c = 3e5  # km/s
H0 = 70.0  # km/s/Mpc
A_cosmo = c / H0

def dL_of_z(z):
    return A_cosmo * z * (1.0 + 0.5 * z)

def z_of_dL(D):
    D_clipped = jnp.maximum(D, 0.0)
    arg = 1.0 + 2.0 * D_clipped / A_cosmo
    arg = jnp.maximum(arg, 1e-12)
    return -1.0 + jnp.sqrt(arg)

def dz_dD(D):
    D_clipped = jnp.maximum(D, 0.0)
    arg = 1.0 + 2.0 * D_clipped / A_cosmo
    arg = jnp.maximum(arg, 1e-12)
    return (1.0 / A_cosmo) / jnp.sqrt(arg)

def chirp_mass(m1, m2):
    M = m1 + m2
    return (m1 * m2)**(3.0/5.0) / (M**(1.0/5.0))

# =========================================================
# True population hyperparameters (MODEL C toy)
# =========================================================

true_alpha_m1   = 2.0
true_mmin       = 5.0
true_mmax       = 60.0
true_lambda_peak = 0.08
true_mpp        = 35.0
true_sig_mpp    = 3.0

true_beta_q     = 1.0
qmin_global     = 0.1

true_kappa_z    = 1.0
zmax_global     = 2.0

true_mu_chi     = 0.0
true_sig_chi    = 0.2

theta_hyper_true = np.array([
    true_alpha_m1, true_mmin, true_mmax,
    true_lambda_peak, true_mpp, true_sig_mpp,
    true_beta_q, true_kappa_z,
    true_mu_chi, np.log(true_sig_chi)
])

# =========================================================
# Reference / injection distribution (MODEL A in source frame)
# =========================================================

alpha_ref   = 1.5
mmin_ref    = 3.0
mmax_ref    = 80.0
qmin_ref    = 0.05
zmax_ref    = 3.0

Z_m1_ref = (mmax_ref**(1.0 - alpha_ref) - mmin_ref**(1.0 - alpha_ref)) / (1.0 - alpha_ref)
Z_q_ref  = 1.0 - qmin_ref
Z_z_ref  = zmax_ref
Z_chi_ref = 2.0  # [-1,1]

def log_p_ref_src(theta_src):
    m1 = theta_src[..., 0]
    q  = theta_src[..., 1]
    z  = theta_src[..., 2]
    chi = theta_src[..., 3]

    mask_m1 = (m1 >= mmin_ref) & (m1 <= mmax_ref)
    m1_clipped = jnp.where(mask_m1, m1, 0.5*(mmin_ref+mmax_ref))
    logp_m1 = -alpha_ref * jnp.log(m1_clipped) - jnp.log(Z_m1_ref)
    logp_m1 = jnp.where(mask_m1, logp_m1, -1e300)

    mask_q = (q >= qmin_ref) & (q <= 1.0)
    logp_q = -jnp.log(Z_q_ref)
    logp_q = jnp.where(mask_q, logp_q, -1e300)

    mask_z = (z >= 0.0) & (z <= zmax_ref)
    logp_z = -jnp.log(Z_z_ref)
    logp_z = jnp.where(mask_z, logp_z, -1e300)

    mask_chi = (chi >= -1.0) & (chi <= 1.0)
    logp_chi = -jnp.log(Z_chi_ref)
    logp_chi = jnp.where(mask_chi, logp_chi, -1e300)

    return logp_m1 + logp_q + logp_z + logp_chi

# =========================================================
# PE prior in DETECTOR frame (for mock PE)
# =========================================================

mu_pe_det_global = np.array([30.0, 20.0, dL_of_z(0.5), 0.0])
sig_pe_det_global = np.array([20.0, 20.0, 1000.0, 1.0])
cov_pe_det_global = np.diag(sig_pe_det_global**2)

# =========================================================
# Transformations: source <-> detector frame
# =========================================================

def src_to_det(theta_src):
    m1, q, z, chi = theta_src[..., 0], theta_src[..., 1], theta_src[..., 2], theta_src[..., 3]
    m2 = q * m1
    m1d = (1.0 + z) * m1
    m2d = (1.0 + z) * m2

    m1d_ord = jnp.maximum(m1d, m2d)
    m2d_ord = jnp.minimum(m1d, m2d)

    dL = dL_of_z(z)
    return jnp.stack([m1d_ord, m2d_ord, dL, chi], axis=-1)

def det_to_src(theta_det):
    m1d, m2d, D, chi = theta_det[..., 0], theta_det[..., 1], theta_det[..., 2], theta_det[..., 3]

    m1d_ord = jnp.maximum(m1d, m2d)
    m2d_ord = jnp.minimum(m1d, m2d)

    D_clipped   = jnp.maximum(D, 1.0)
    m1d_clipped = jnp.maximum(m1d_ord, 1e-3)

    z = z_of_dL(D_clipped)
    m1 = m1d_clipped / (1.0 + z)
    q  = m2d_ord / m1d_clipped

    return jnp.stack([m1, q, z, chi], axis=-1)

def log_abs_detJ_det_to_src(theta_det):
    m1d, m2d, D, chi = theta_det[..., 0], theta_det[..., 1], theta_det[..., 2], theta_det[..., 3]

    m1d_ord = jnp.maximum(m1d, m2d)
    D_clipped   = jnp.maximum(D, 1.0)
    m1d_clipped = jnp.maximum(m1d_ord, 1e-3)

    z    = z_of_dL(D_clipped)
    dzdD = dz_dD(D_clipped)

    J = jnp.abs(dzdD / ((1.0 + z) * m1d_clipped))
    return jnp.log(jnp.clip(J, 1e-300, jnp.inf))

# =========================================================
# GW-ish SNR model and detection (detector frame)
# =========================================================

def snr_det(theta_det, K=1000.0):
    m1d, m2d, D, chi = theta_det[..., 0], theta_det[..., 1], theta_det[..., 2], theta_det[..., 3]
    Mc_det = chirp_mass(m1d, m2d)
    D = jnp.maximum(D, 1.0)
    rho = K * (Mc_det**(5.0/6.0)) / D
    return rho

def p_snr(theta_det, rho_thr=2.0, sigma_rho=1.0):
    rho = snr_det(theta_det)
    z = (rho_thr - rho) / sigma_rho
    return 0.5 * (1.0 - jsp.special.erf(z / jnp.sqrt(2.0)))

def p_cut(theta_det, mc_max=100.0):
    theta_src = det_to_src(theta_det)
    m1 = theta_src[..., 0]
    q  = theta_src[..., 1]
    m2 = q * m1
    Mc_src = chirp_mass(m1, m2)
    return (Mc_src < mc_max).astype(float)

def p_det(theta_det):
    return p_snr(theta_det) * p_cut(theta_det)

# =========================================================
# MODEL C population in source frame
# =========================================================

def log_p_m1(m1, alpha, mmin, mmax, lambda_peak, mpp, sig_mpp):
    mask = (m1 >= mmin) & (m1 <= mmax)
    m1_clipped = jnp.where(mask, m1, 0.5*(mmin+mmax))

    pl = (1.0 - lambda_peak) * (m1_clipped**(-alpha))
    pg = lambda_peak * jnp.exp(-0.5*((m1_clipped - mpp)/sig_mpp)**2) / (sig_mpp * jnp.sqrt(2.0*jnp.pi))
    f = pl + pg

    grid = jnp.linspace(mmin, mmax, 512)
    pl_g = (1.0 - lambda_peak) * (grid**(-alpha))
    pg_g = lambda_peak * jnp.exp(-0.5*((grid - mpp)/sig_mpp)**2) / (sig_mpp * jnp.sqrt(2.0*jnp.pi))
    f_g = pl_g + pg_g
    Z = jnp.trapezoid(f_g, grid)
    Z = jnp.clip(Z, 1e-300, jnp.inf)

    logp = jnp.log(jnp.clip(f, 1e-300, jnp.inf)) - jnp.log(Z)
    logp = jnp.where(mask, logp, -1e300)
    return logp

def log_p_q(q, beta_q, qmin=0.1):
    mask = (q >= qmin) & (q <= 1.0)
    q_clipped = jnp.where(mask, q, 0.5*(qmin+1.0))

    Z_general = (1.0 - qmin**(beta_q+1.0)) / (beta_q+1.0)
    Z_special = jnp.log(1.0/qmin)
    Z = jnp.where(jnp.abs(beta_q + 1.0) < 1e-6, Z_special, Z_general)
    Z = jnp.clip(Z, 1e-300, jnp.inf)

    logp = beta_q * jnp.log(q_clipped) - jnp.log(Z)
    logp = jnp.where(mask, logp, -1e300)
    return logp

def log_p_z(z, kappa_z, zmax=2.0):
    mask = (z >= 0.0) & (z <= zmax)
    z_clipped = jnp.where(mask, z, 0.5*zmax)

    f = (z_clipped**2) * ((1.0+z_clipped)**(kappa_z-1.0))

    grid = jnp.linspace(0.0, zmax, 512)
    f_g = (grid**2) * ((1.0+grid)**(kappa_z-1.0))
    Z = jnp.trapezoid(f_g, grid)
    Z = jnp.clip(Z, 1e-300, jnp.inf)

    logp = jnp.log(jnp.clip(f, 1e-300, jnp.inf)) - jnp.log(Z)
    logp = jnp.where(mask, logp, -1e300)
    return logp

def log_p_chi(chi, mu_chi, sig_chi):
    mask = (chi >= -1.0) & (chi <= 1.0)
    chi_clipped = jnp.where(mask, chi, 0.0)

    f = jnp.exp(-0.5*((chi_clipped - mu_chi)/sig_chi)**2) / (sig_chi * jnp.sqrt(2.0*jnp.pi))

    grid = jnp.linspace(-1.0, 1.0, 512)
    f_g = jnp.exp(-0.5*((grid - mu_chi)/sig_chi)**2) / (sig_chi * jnp.sqrt(2.0*jnp.pi))
    Z = jnp.trapezoid(f_g, grid)
    Z = jnp.clip(Z, 1e-300, jnp.inf)

    logp = jnp.log(jnp.clip(f, 1e-300, jnp.inf)) - jnp.log(Z)
    logp = jnp.where(mask, logp, -1e300)
    return logp

def log_p_pop_src(theta_src, theta_hyper):
    alpha, mmin, mmax, lambda_peak, mpp, sig_mpp, beta_q, kappa_z, mu_chi, log_sig_chi = theta_hyper
    sig_chi = jnp.exp(log_sig_chi)

    m1 = theta_src[..., 0]
    q  = theta_src[..., 1]
    z  = theta_src[..., 2]
    chi = theta_src[..., 3]

    lp_m1  = log_p_m1(m1, alpha, mmin, mmax, lambda_peak, mpp, sig_mpp)
    lp_q   = log_p_q(q, beta_q, qmin_global)
    lp_z   = log_p_z(z, kappa_z, zmax_global)
    lp_chi = log_p_chi(chi, mu_chi, sig_chi)

    return lp_m1 + lp_q + lp_z + lp_chi

# =========================================================
# Sampling from true population (MODEL C)
# =========================================================

def sample_m1_from_true(rng, n):
    alpha = true_alpha_m1
    mmin = true_mmin
    mmax = true_mmax
    u = rng.uniform(size=n)
    if alpha != 1.0:
        return (mmin**(1.0-alpha) + u*(mmax**(1.0-alpha) - mmin**(1.0-alpha)))**(1.0/(1.0-alpha))
    else:
        return mmin * (mmax/mmin)**u

def sample_q_from_true(rng, n):
    beta_q = true_beta_q
    qmin = qmin_global
    u = rng.uniform(size=n)
    if beta_q != -1.0:
        return (qmin**(beta_q+1.0) + u*(1.0 - qmin**(beta_q+1.0)))**(1.0/(beta_q+1.0))
    else:
        return qmin**u

def sample_z_from_true(rng, n):
    kappa_z = true_kappa_z
    zmax = zmax_global
    samples = []
    needed = n
    while needed > 0:
        z_trial = rng.uniform(0.0, zmax, size=needed*4)
        f = (z_trial**2) * ((1.0+z_trial)**(kappa_z-1.0))
        fmax = f.max()
        u = rng.uniform(0, fmax, size=f.size)
        accepted = z_trial[u < f]
        if accepted.size > 0:
            take = min(accepted.size, needed)
            samples.append(accepted[:take])
            needed -= take
    return np.concatenate(samples)

def sample_chi_from_true(rng, n):
    mu_chi, sig_chi = true_mu_chi, true_sig_chi
    samples = []
    needed = n
    while needed > 0:
        chi_trial = rng.normal(mu_chi, sig_chi, size=needed*4)
        mask = (chi_trial >= -1.0) & (chi_trial <= 1.0)
        accepted = chi_trial[mask]
        if accepted.size > 0:
            take = min(accepted.size, needed)
            samples.append(accepted[:take])
            needed -= take
    return np.concatenate(samples)

# =========================================================
# Reference MODEL A sampling (for injections)
# =========================================================

def sample_m1_ref(rng, n):
    alpha = alpha_ref
    mmin = mmin_ref
    mmax = mmax_ref
    u = rng.uniform(size=n)
    if alpha != 1.0:
        return (mmin**(1.0-alpha) + u*(mmax**(1.0-alpha) - mmin**(1.0-alpha)))**(1.0/(1.0-alpha))
    else:
        return mmin * (mmax/mmin)**u

def sample_q_ref(rng, n):
    qmin = qmin_ref
    u = rng.uniform(size=n)
    return qmin + (1.0 - qmin)*u

def sample_z_ref(rng, n):
    zmax = zmax_ref
    return rng.uniform(0.0, zmax, size=n)

def sample_chi_ref(rng, n):
    return rng.uniform(-1.0, 1.0, size=n)

# =========================================================
# Posterior sampling for mock PE
# =========================================================

def posterior_params_pe(mu_pe_det, cov_pe_det, cov_obs_det, data_det):
    cov_post = np.linalg.inv(np.linalg.inv(cov_pe_det) + np.linalg.inv(cov_obs_det))
    mu_post = cov_post @ (np.linalg.inv(cov_pe_det) @ mu_pe_det + np.linalg.inv(cov_obs_det) @ data_det)
    return mu_post, cov_post

def sample_posterior(rng, mu_post, cov_post, nsamples):
    raw = rng.multivariate_normal(mu_post, cov_post, size=nsamples)

    m1d = raw[:, 0]
    m2d = raw[:, 1]
    dL  = raw[:, 2]
    chi = raw[:, 3]

    m1d_pos = np.maximum(m1d, 1e-3)
    m2d_pos = np.maximum(m2d, 1e-3)
    m1d_ord = np.maximum(m1d_pos, m2d_pos)
    m2d_ord = np.minimum(m1d_pos, m2d_pos)

    dL_pos = np.maximum(dL, 1.0)
    chi_clip = np.clip(chi, -1.0, 1.0)

    return np.stack([m1d_ord, m2d_ord, dL_pos, chi_clip], axis=-1)

# =========================================================
# Simulate dataset
# =========================================================

def simulate_dataset(seed,
                     cov_obs_det=np.diag([1.0, 1.0, 10.0, 0.05]),
                     R_true=300.0,
                     nsamples=4000,
                     n_inj=int(1e5)):
    rng = np.random.default_rng(seed)

    mu_pe_det = mu_pe_det_global
    cov_pe_det = cov_pe_det_global

    N_tot = rng.poisson(R_true)

    m1 = sample_m1_from_true(rng, N_tot)
    q  = sample_q_from_true(rng, N_tot)
    z  = sample_z_from_true(rng, N_tot)
    chi = sample_chi_from_true(rng, N_tot)
    thetas_src = np.vstack([m1, q, z, chi]).T

    thetas_det = np.array(src_to_det(jnp.array(thetas_src)))

    p_det_np = np.array(p_det(jnp.array(thetas_det)))
    u = rng.uniform(size=N_tot)
    detected = u < p_det_np
    thetas_det_obs = thetas_det[detected]
    thetas_src_obs = thetas_src[detected]
    N_obs = len(thetas_det_obs)

    data_det = np.empty_like(thetas_det_obs)
    for k in range(N_obs):
        data_det[k] = rng.multivariate_normal(thetas_det_obs[k], cov_obs_det)

    samples_det = np.empty((N_obs, nsamples, 4))
    for k in range(N_obs):
        mu_post, cov_post = posterior_params_pe(mu_pe_det, cov_pe_det, cov_obs_det, data_det[k])
        samples_det[k] = sample_posterior(rng, mu_post, cov_post, nsamples)

    m1_ref = sample_m1_ref(rng, n_inj)
    q_ref  = sample_q_ref(rng, n_inj)
    z_ref  = sample_z_ref(rng, n_inj)
    chi_ref = sample_chi_ref(rng, n_inj)
    theta_src_ref = np.vstack([m1_ref, q_ref, z_ref, chi_ref]).T
    injections_det = np.array(src_to_det(jnp.array(theta_src_ref)))

    print("Total simulated:", N_tot)
    print("Detected:", N_obs)

    return injections_det, samples_det, data_det, theta_hyper_true, cov_obs_det, mu_pe_det, cov_pe_det

# =========================================================
# Train one flow per event (source-frame)
# =========================================================

def train_flow_for_event(theta_src_event, key, n_epochs=50, batch_size=128):
    dim = theta_src_event.shape[-1]
    base_dist = Normal(jnp.zeros(dim), jnp.ones(dim))

    flow = masked_autoregressive_flow(
        key=key,
        base_dist=base_dist,
        flow_layers=4,
        nn_width=64,
        nn_depth=2,
    )

    optimizer = optax.adam(1e-3)
    opt_state = optimizer.init(eqx.filter(flow, eqx.is_array))

    @eqx.filter_value_and_grad
    def loss_fn(flow, x):
        return -jnp.mean(flow.log_prob(x))

    @eqx.filter_jit
    def train_step(flow, opt_state, x):
        loss, grads = loss_fn(flow, x)
        updates, opt_state = optimizer.update(grads, opt_state)
        flow = eqx.apply_updates(flow, updates)
        return flow, opt_state, loss

    nsamples = theta_src_event.shape[0]

    for epoch in range(1, n_epochs + 1):
        perm = np.random.permutation(nsamples)
        x_perm = theta_src_event[perm]

        for i in range(0, nsamples, batch_size):
            batch = x_perm[i:i+batch_size]
            if batch.shape[0] < batch_size:
                continue
            flow, opt_state, loss = train_step(flow, opt_state, batch)

    return flow

# =========================================================
# Hierarchical likelihood with per-event flows
# =========================================================

def make_loglikelihood(injections_det_jax, samples_det_jax, flows_per_event):
    nsel_jax = injections_det_jax.shape[0]
    nobs_jax, nsamples_jax = samples_det_jax.shape[:2]

    @jax.jit
    def loglikelihood(theta_hyper):
        theta_det_inj = injections_det_jax
        theta_src_inj = det_to_src(theta_det_inj)
        logJ_inj = log_abs_detJ_det_to_src(theta_det_inj)

        log_p_pop_inj_src = log_p_pop_src(theta_src_inj, theta_hyper)
        log_p_pop_inj_det = log_p_pop_inj_src + logJ_inj

        log_p_ref_inj_src = log_p_ref_src(theta_src_inj)
        log_p_ref_inj_det = log_p_ref_inj_src + logJ_inj

        logw_inj = log_p_pop_inj_det - log_p_ref_inj_det
        pdet_inj = jnp.clip(p_det(theta_det_inj), 1e-300, 1.0)
        logw_inj_det = logw_inj + jnp.log(pdet_inj)

        log_alpha = jsp.special.logsumexp(logw_inj_det) - jnp.log(nsel_jax)

        log_Zk_list = []

        for k in range(nobs_jax):
            theta_det_k = samples_det_jax[k]          # (nsamples, 4)
            theta_src_k = det_to_src(theta_det_k)     # (nsamples, 4)
            logJ_k      = log_abs_detJ_det_to_src(theta_det_k)

            log_p_pop_src_k = log_p_pop_src(theta_src_k, theta_hyper)
            log_p_pop_det_k = log_p_pop_src_k + logJ_k

            log_pi_pe_k = flows_per_event[k].log_prob(theta_src_k)  # (nsamples,)
            log_pdet_k  = jnp.log(jnp.clip(p_det(theta_det_k), 1e-300, 1.0))

            logw_k = log_p_pop_det_k + log_pdet_k - log_pi_pe_k
            log_Zk = jsp.special.logsumexp(logw_k) - jnp.log(nsamples_jax)
            log_Zk_list.append(log_Zk)

        log_Zk_all = jnp.stack(log_Zk_list)

        return jnp.sum(log_Zk_all) - nobs_jax * log_alpha

    return loglikelihood

# =========================================================
# Dynesty prior
# =========================================================

def prior_transform(u):
    alpha_low, alpha_high = 1.0, 5.0
    mmin_low, mmin_high = 3.0, 10.0
    mmax_low, mmax_high = 50.0, 100.0
    lambda_low, lambda_high = 0.0, 1.0
    sigpp_low, sigpp_high = 1.0, 5.0
    betaq_low, betaq_high = -1.0, 3.0
    kappaz_low, kappaz_high = -2.0, 4.0
    muchi_low, muchi_high = -0.5, 0.5
    logsigchi_low, logsigchi_high = np.log(0.05), np.log(0.5)

    alpha = alpha_low + (alpha_high - alpha_low) * u[0]

    mmin = mmin_low + (mmin_high - mmin_low) * u[1]
    mmax = mmax_low + (mmax_high - mmax_low) * u[2]
    if mmax <= mmin:
        mmax = mmin + 1e-3

    lambda_peak = lambda_low + (lambda_high - lambda_low) * u[3]
    mpp = mmin + (mmax - mmin) * u[4]
    sig_mpp = sigpp_low + (sigpp_high - sigpp_low) * u[5]

    beta_q  = betaq_low  + (betaq_high  - betaq_low)  * u[6]
    kappa_z = kappaz_low + (kappaz_high - kappaz_low) * u[7]

    mu_chi      = muchi_low + (muchi_high - muchi_low) * u[8]
    log_sig_chi = logsigchi_low + (logsigchi_high - logsigchi_low) * u[9]

    return np.array([
        alpha, mmin, mmax,
        lambda_peak, mpp, sig_mpp,
        beta_q, kappa_z,
        mu_chi, log_sig_chi
    ])

# =========================================================
# Full pipeline
# =========================================================

def run_full_pipeline(seed=0):
    injections_det, samples_det, data_det, theta_hyper_true, cov_obs_det, mu_pe_det, cov_pe_det = simulate_dataset(seed)

    injections_det_jax = jnp.array(injections_det)
    samples_det_jax    = jnp.array(samples_det)

    N_obs = samples_det_jax.shape[0]
    nsamples = samples_det_jax.shape[1]

    samples_src_jax = det_to_src(samples_det_jax)  # (N_obs, nsamples, 4)

    flows_per_event = []
    key = jax.random.key(0)

    for k in range(N_obs):
        key, subkey = jax.random.split(key)
        theta_src_k = samples_src_jax[k]           # (nsamples, 4)
        flow_k = train_flow_for_event(theta_src_k, subkey, n_epochs=40, batch_size=1000)
        flows_per_event.append(flow_k)
        print(f"Trained flow for event {k+1}/{N_obs}")

    loglikelihood_jax = make_loglikelihood(
        injections_det_jax,
        samples_det_jax,
        flows_per_event
    )

    def loglikelihood_np(theta_hyper):
        return float(loglikelihood_jax(jnp.asarray(theta_hyper)))

    sampler = NestedSampler(loglikelihood_np, prior_transform, ndim=10, nlive=400)
    sampler.run_nested(print_progress=True)
    res = sampler.results

    weights = np.exp(res.logwt - res.logz[-1])
    samples_hyper = dynesty.utils.resample_equal(res.samples, weights)

    mean_post = samples_hyper.mean(axis=0)
    std_post  = samples_hyper.std(axis=0)

    print("\n==================== FLOW-PER-EVENT HIERARCHICAL RESULTS ====================")
    print("Posterior mean:")
    print(mean_post)
    print("Posterior std:")
    print(std_post)
    print("True hyperparams:")
    print(theta_hyper_true)
    print("============================================================================\n")

    fig = corner.corner(
        samples_hyper[:, :6],
        labels=[r"$\alpha$", r"$m_{\min}$", r"$m_{\max}$",
                r"$\lambda_{\rm peak}$", r"$m_{\rm pp}$", r"$\sigma_{\rm pp}$"],
        truths=theta_hyper_true[:6],
        truth_color="red",
        show_titles=True,
        title_fmt=".3f",
        quantiles=[0.16, 0.5, 0.84],
        levels=[0.68, 0.95],
        fill_contours=True,
        smooth=1.0,
    )
    plt.show()

    return samples_hyper

if __name__ == "__main__":
    samples_hyper = run_full_pipeline(seed=0)
```
